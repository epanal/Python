## Problem Understanding
Credit card defaults significantly impact banks, as seen with Taiwanese banks around 2005 when credit card debt ballooned to $268 billion USD. Credit delinquency leads to financial losses for a bank and affects its overall profitability. Secondary effects of these losses include reputational damage, loss of customer trust, higher regulatory scrutiny, and increased resources and legal efforts to collect money owed from defaulted accounts.

By predicting potential credit card defaults, banks can mitigate risk through identifying early on which credit card users may be high risk or setting appropriate interest rates based on a customer's assessed risk.
